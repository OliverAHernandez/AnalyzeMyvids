```python
# I turned the dates into string objects to make it easier to manipulate
df['trending_date'] = df['trending_date'].map(lambda x: str(x).split(' ')[0][2:])
```

```python
# created this function to realign the months,days,years
def conv_date(x):
    first,second,last = x.split('-')
    this = None
    try:
        this = pd.to_datetime('20' + last + '-' +second +'-' + first)
        return this
    except:
        pass
    try:
        this = pd.to_datetime('20' + first + '-' +last +'-' + second)
        return this
    except:
        pass
    try:
        this = pd.to_datetime('20' + last + '-' +first +'-' + second)
        return this
    except:
        pass
    try:
        this = pd.to_datetime('20' + second + '-' +first +'-' + last)
        return this
    except:
        pass
    try:
        this = pd.to_datetime('20' + first + '-' +second +'-' + last)
        return this
    except:
        pass
    try:
        this = pd.to_datetime('20' + second + '-' +last +'-' + first)
        return this
    except:
        pass
    else:
        return x
```

```python
df['trending_date'] = df.trending_date.map(conv_date)
```
